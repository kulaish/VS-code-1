Read the mall customer datset

```python
import pandas as pd
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.columns = ['CustomerID', 'Gender', 'Age', 'Income', 'Spend']
```

```python
df.head()
```

Seperative X data

```python
X = df[['Age', 'Income', 'Spend']]
X.head()
```

Preprocess X data

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

```python
X_pre.mean()
```

```python
X_pre.std()
```

Calculate optimal number of clusters on the basis of Silhoutee score

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
n = list(range(2, 15))
n
```

```python
scores = []
for i in n:
    # Create the model
    model = AgglomerativeClustering(n_clusters=i)
    model.fit(X_pre)
    # Predict the model clusters
    lb = model.labels_
    # calculate silhoutee scores
    sil = silhouette_score(X_pre, lb)
    scores.append(sil)
```

```python
scores
```

Visualize silhoutee scores

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(12, 6))
plt.plot(n, scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhoutee Score')
plt.title('Silhoutee Scores for various clusters')
plt.grid()
plt.show()
```

Based on above graph for 6 clusters we have highest score of 0.42

Create a Agglomerative model with n_clusters=6

```python
model = AgglomerativeClustering(n_clusters=6)
model.fit(X_pre)
```

```python
model.labels_
```

Save above results in X

```python
X['Label']=model.labels_
```
